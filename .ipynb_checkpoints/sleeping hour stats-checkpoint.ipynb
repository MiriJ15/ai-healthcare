```python
import pandas as pd
import matplotlib.pyplot as plt

sleep = pd.read_csv("C:/Users/hwenn/ai-healthcare/sleep.csv", delimiter=r'\s+')  # For whitespace
```

```python
pre = sleep[sleep['date'] <= 20200311]
during = sleep[(sleep['date'] > 20200311)]
print(len(pre), len(during))
print(sleep['id'].nunique())
```

```python
# Convert 'date' column to datetime
sleep['date'] = pd.to_datetime(sleep['date'], format='%Y%m%d')

# Group by 'date' and 'id', then calculate the mean sleep hours per day across unique ids
daily_sleep = sleep.groupby(['date', 'id'])['sleep_duration_hours'].mean().reset_index()

# Calculate the daily mean and standard deviation of sleep duration
sleep_stats = daily_sleep.groupby('date')['sleep_duration_hours'].agg(['mean', 'std']).reset_index()

# Split date for pre and during periods
date_split = pd.to_datetime("2020-03-11")
# Plot
plt.figure(figsize=(12, 6))

# Plot the average sleep duration over time
plt.plot(sleep_stats['date'], sleep_stats['mean'], label="Average Sleep Hours")

# Shade the standard deviation around the mean
plt.fill_between(
    sleep_stats['date'],
    sleep_stats['mean'] - sleep_stats['std'],
    sleep_stats['mean'] + sleep_stats['std'],
    alpha=0.3,
    label="Standard Deviation"
)

# Add vertical line to indicate the split
plt.axvline(date_split, color='red', linestyle='--', linewidth=1.5, label="Split Date: 2020-03-11")

# Labels and legend
plt.title("Average Sleep Hours Over Time with Standard Deviation")
plt.xlabel("Date")
plt.ylabel("Average Sleep Hours")
plt.legend()

plt.show()
```

```python
peak_days = sleep_stats.sort_values(by='mean', ascending=False).head(5)
print("Top 5 Peak Days by Average Sleep Duration:")
print(peak_days[['date', 'mean']])
```

```python
# Define the cutoff date
cutoff_date = pd.to_datetime("2023-05-06")

# Filter entries after the cutoff date
entries_after_cutoff = sleep[sleep['date'] > cutoff_date]

# Count the number of entries after the cutoff date
num_entries_after_cutoff = len(entries_after_cutoff)
print(f"Number of entries after 2023-05-06: {num_entries_after_cutoff}")

# Find the latest date of entry in the dataset
latest_date = sleep['date'].max()
print(f"Latest date of entry: {latest_date}")
```

```python
zero_sleep_duration_count = (sleep['sleep_duration_hours'] == 0).sum()
zero_sleep_heathkit_count = (sleep['sleep_heathkit_dur'] == 0).sum()
print(zero_sleep_duration_count, zero_sleep_heathkit_count)

# Replace 0 with 0.1 in 'sleep_duration_hours' and 'sleep_heathkit_dur'
sleep['sleep_duration_hours'] = sleep['sleep_duration_hours'].replace(0, 0.1)
sleep['sleep_heathkit_dur'] = sleep['sleep_heathkit_dur'].replace(0, 0.1)
```

```python
# Step 1: Calculate fold difference and absolute hour difference, ignoring NaNs
# Use where conditions to avoid calculations on rows with NaNs
sleep['fold_difference'] = sleep.apply(
    lambda row: max(row['sleep_duration_hours'], row['sleep_heathkit_dur']) / min(row['sleep_duration_hours'], row['sleep_heathkit_dur']) 
    if pd.notna(row['sleep_duration_hours']) and pd.notna(row['sleep_heathkit_dur']) else None, axis=1
)

sleep['hour_difference'] = sleep.apply(
    lambda row: abs(row['sleep_duration_hours'] - row['sleep_heathkit_dur']) 
    if pd.notna(row['sleep_duration_hours']) and pd.notna(row['sleep_heathkit_dur']) else None, axis=1
)

# Step 2: Filter out rows based on fold difference and hour difference criteria
# Retain rows where fold difference is <= 1.5 or hour difference is <= 2,
# or if either value is NaN (indicating no calculations were done)
filtered_sleep = sleep[
    (sleep['fold_difference'].isna() | (sleep['fold_difference'] <= 1.5)) &
    (sleep['hour_difference'].isna() | (sleep['hour_difference'] <= 2))
]

# Display the filtered dataset
print(f"Original number of rows: {len(sleep)}")
print(f"Number of rows after filtering: {len(filtered_sleep)}")
```

```python
filtered_daily_sleep = filtered_sleep.groupby(['date', 'id'])['sleep_duration_hours'].mean().reset_index()

# Calculate daily mean and standard deviation across all students
filtered_sleep_stats = filtered_daily_sleep.groupby('date')['sleep_duration_hours'].agg(['mean', 'std']).reset_index()

# Filter out days with average sleep hours < 6 or > 9
filtered_sleep_stats = filtered_sleep_stats[(sleep_stats['mean'] >= 6) & (sleep_stats['mean'] <= 9)]
```

```python
# Step 2: Plot the line plot with standard deviation shading
plt.figure(figsize=(8, 4))

# Line plot of mean sleep duration
plt.plot(filtered_sleep_stats['date'], filtered_sleep_stats['mean'], label="Average Sleep Hours")

# Shaded area for standard deviation
plt.fill_between(
    filtered_sleep_stats['date'],
    filtered_sleep_stats['mean'] - filtered_sleep_stats['std'],
    filtered_sleep_stats['mean'] + filtered_sleep_stats['std'],
    alpha=0.3,
    label="Standard Deviation"
)

plt.axvline(date_split, color='red', linestyle='--', linewidth=1.5, label="Split Date: 2020-03-11")

# Labels and legend
plt.title("Average Student Sleep Duration Before and During Covid")
plt.xlabel("Date")
plt.ylabel("Hours")
plt.legend()

plt.tight_layout()
plt.show()
```
